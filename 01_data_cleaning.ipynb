## 2] Orders_Dataset

### *~ Loading Dataset ~*

```python
import pandas as pd

orders = pd.read_csv(r"C:\Users\HOLE\DATA ANALYST\RetailPulseAI - Project\01_Data\Raw\olist_orders_dataset.csv")
orders.head()
```

#### *~ Data Overview ~*

```python
print("=" * 60)
print("\t\t📦 ORDERS DATASET OVERVIEW")
print("=" * 60)

print("\n📌 Shape of the Dataset")
print(orders.shape)

print("\n📌 Total Orders")
print(orders.shape[0])

print("\n📌 Dataset Information")
orders.info()

print("\n📌 Unique Values in Each Column")
print(f"Total Columns: {len(orders.columns)}")
print(orders.nunique())
```

```python
print("=" * 60)
print("\t\t🔄 DATATYPE CONVERSION")
print("=" * 60)

date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")

print("\n📌 Date Columns Converted Successfully")
print(date_cols)

print("\n📌 Updated Data Types")
print(orders[date_cols].dtypes)
```

#### *~ Data Cleaning ~*

```python
print("=" * 60)
print("\t\t🧹 DATA CLEANING REPORT")
print("=" * 60)

print(f"\n📊 Dataset Shape : {orders.shape}")

print("\n📌 Missing Values")
print("-" * 60)
print(orders.isnull().sum())

print("\n📌 Duplicate Records")
print("-" * 60)
print(f"Total Duplicate Rows : {orders.duplicated().sum()}")

print("\n📌 Data Type Validation")
print("-" * 60)
print(orders.dtypes)
```

```python
print("=" * 60)
print("\t\t🧹 DATA CLEANING")
print("=" * 60)

print("\n📌 Missing Values Before Cleaning")
print(orders.isnull().sum())

# Handle missing values
orders["order_approved_at"] = orders["order_approved_at"].fillna(
    orders["order_purchase_timestamp"]
)

orders["order_delivered_carrier_date"] = orders["order_delivered_carrier_date"].fillna(
    orders["order_approved_at"]
)

orders["order_delivered_customer_date"] = orders["order_delivered_customer_date"].fillna(
    orders["order_estimated_delivery_date"]
)

print("\n📌 Missing Values After Cleaning")
print(orders.isnull().sum())
```
